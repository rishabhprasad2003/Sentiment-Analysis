```python
webpage = requests.get('https://www.airlinequality.com/airline-reviews/british-airways/page/1/').text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
review_title = []
name = []
date = []
rating = []
verification = []
review_text = []
```

```python
final = pd.DataFrame()

for i in range(1,361):
    
    webpage = requests.get(f'https://www.airlinequality.com/airline-reviews/british-airways/page/{i}/').text
    
    soup = BeautifulSoup(webpage, 'lxml')
    
    review_body = soup.find_all('div', class_='body')
    k = 0
    for j in review_body:
        review_title.append(j.find('h2',class_='text_header').text)
        
        name.append(j.find('span',itemprop='name').text)
        #name.append(j.select('span[itemprop="name"]')[k].text)
        
        try:
            date.append(j.find('time', itemprop='datePublished').text)
        except:
            date.append(np.nan)
        
        # resolve
        try:
            rating.append(soup.find_all('span', itemprop='ratingValue')[k].text.strip())
        except:
            rating.append(np.nan)
        
        try:
            verification.append(j.find('em').text)
        except:
            verification.append(np.nan)
        
        review_text.append(j.find('div',class_='text_content').text)
        k += 1
```

```python
d = {'review_title':review_title, 'name':name, 'date':date, 'rating':rating, 'verification':verification, 'review_text':review_text}

df = pd.DataFrame(d)

final = final.append(df)
```
